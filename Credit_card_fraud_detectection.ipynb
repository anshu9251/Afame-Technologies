```python
df = df.drop_duplicates(keep="first")
```

```python
#Distribution of data bteween legit and fraud transaction
```

```python
legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit.describe()
```

```python
fraud.describe()
```

```python
df.groupby("Class").mean()
```

```python
#Performing Undersampling to deal with imbalance data
```

```python
legit_sample = legit.sample(n=473)
```

```python
legit_sample.shape
```

```python
df_new = pd.concat([legit_sample,fraud],axis=0)
```

```python
df_new.shape
```

```python
df_new.head()
```

```python
df_new.tail()
```

```python
df_new["Class"].value_counts()
```

```python
df_new.groupby("Class").mean()
```

```python
#Splitting our dataset into features and target
```

```python
X = df_new.drop(columns="Class",axis=1)
y = df_new["Class"]
```

```python
X.shape
```

```python
X
```

```python
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Model Building
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
#accuracy score on training data
```

```python
X_train_pred = model.predict(X_train)
```

```python
X_train_pred.shape
```

```python
accuracy_score(X_train_pred,y_train)
```

```python
#accuracy score on testing data
```

```python
y_pred = model.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
rf = RandomForestClassifier(n_estimators=150)
etc = ExtraTreesClassifier(n_estimators=100)
```

```python
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)

print(accuracy_score(y_pred,y_test))
```

```python
etc.fit(X_train,y_train)
y_pred = etc.predict(X_test)

print(accuracy_score(y_pred,y_test))
```
